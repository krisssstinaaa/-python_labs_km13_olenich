Розглянемо конкретний приклад.

Припустимо, що кожен зшитий зошит нашої книги складається з 8 листів (16 сторінок). Якщо уявити великі листи, з яких потім складаються зошити, то на кожному таком листі буде по 4 сторінки книги. У нашій книзі таких великих листів буде 4. Розглянемо перший лист. З одного боку на ньому будуть надруковані сторінки 16 та 1, з іншого - 2 та 15. **Написати програму, яка визначатиме послідовність сторінок, що будуть надруковані на кожному великому листі.**

4. Створіть функцію, яка прийматиме на вхід число, введене користувачем, і повертатиме список із вкладеними списками. Кожен вкладений список має зберігати номери сторінок для кожного зошита у потрібній послідовності. Для 16-сторінкового зошита список повинен мати вигляд:

Якщо параметр дорівнює False, додаткові функції декоратора не повинні виконуватись, тобто має повертатись список з п. 4. Якщо параметр дорівнює True, декоратор повинен виділяти кожні 4 номери сторінок (тобто сторінки, надруковані на одному великому листі), в окремі кортежі або списки. 6. Замість функції з п. 4 створіть генератор. Функція-декоратор має працювати з цим генератором так само, як і зі звичайною функцією.

```python

def split_tuple(numbers):
    return [tuple(numbers[0:4]), tuple(numbers[4:8]), tuple(numbers[8:12]), tuple(numbers[12:])]


def decorator(split_list):
    def wrap(function):
        def wrapper(number):
            if split_list:
                return [*map(split_tuple, function(number))]
            else:
                return function(number)
        return wrapper
    return wrap


@decorator(True)
def book(number):
    assert isinstance(number, int) and number > 15 and number < 1281 and number % 16 == 0 
    result = []
    for i in range(number // 16):
        sheets = []
        step = i * 16
        for j in range(1,9):
            sheets += [j + step, 16 - j + 1 + step] if j % 2 == 0 else [16 - j + 1 + step, j + step]
        result.append(sheets)

    return result


def my_input():
    while True:
        a = input("Enter the number, please: ")

        try:
            number = int(a)
            if number > 15 and number < 1281 and number % 16 == 0: return number
            else: print('Error: 16 <= number <= 1280, num mod16 = 0, try again')
        except ValueError as error:
            print(f'Error: {error}, try again')

number = my_input()

print(book(number))
```
